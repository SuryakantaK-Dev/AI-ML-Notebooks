#Step 2: Import Necessary Modules
Import FACTSCORE and other required libraries:

```python
from factscore import FactScorer
```

#Step 3: Prepare Sample Input

```python
# Example long-form text
generated_text = """
Bridget Moynahan is an American actress, model, and producer.
She is best known for her roles in Grey’s Anatomy and Blue Bloods.
"""

# Example trusted source (Wikipedia-style data)
knowledge_source = """
Bridget Moynahan is an American actress. She is best known for her role in Blue Bloods.
She studied acting at the American Academy of Dramatic Arts.
"""
```

#Step 4: Break Text into Atomic Facts

```python
scorer = FactScorer()
atomic_facts = scorer.extract_atomic_facts(generated_text)
print("Atomic Facts:", atomic_facts)
```

#Step 5: Evaluate Factual Precision
Validate the atomic facts against the knowledge source:

```python
factscore, validated_facts = scorer.evaluate(atomic_facts, knowledge_source)
print("FACTSCORE:", factscore)
print("Validated Facts:", validated_facts)
```
