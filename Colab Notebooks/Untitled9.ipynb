```python
# Import pandas
import pandas as pd

# Read the CSV file
df = pd.read_csv("NER_Dataset.csv")

# Group the sentences and tags by the sentence ID
sentences = df.groupby("Sentence_ID")["Word"].apply(list).values
tags = df.groupby("Sentence_ID")["Tag"].apply(list).values

# Load the TAG Mapping.json file
import json
with open("TAG_Mapping.json") as f:
  tag_map = json.load(f)

# Map the dataset tags to the model tags
tags = [[tag_map[t] for t in tag_seq] for tag_seq in tags]
```
