```python
import os
import cv2
import numpy as np
from sklearn.metrics import f1_score, roc_auc_score
import os
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
os.chdir('/content/drive/My Drive/surya/ML_Assignment/gurmukhi/')
# Load the training data
train_dir = '/content/drive/MyDrive/surya/NCVPRIPG-writer-verification-main/dataset/train'
train_data = []
```

```python

for folder in os.listdir(train_dir):
    folder_path = os.path.join(train_dir, folder)
    for file_name in os.listdir(folder_path):
        image_path = os.path.join(folder_path, file_name)
        train_data.append((image_path, folder))

# Split the training data into train and validation sets
train_data, val_data = train_test_split(train_data, test_size=0.2, random_state=42)
```

```python
# Load the validation data
val_df = pd.read_csv('/content/drive/MyDrive/surya/NCVPRIPG-writer-verification-main/dataset/val.csv')
```

```python
# Preprocess the data
def preprocess_data(data):
    processed_data = []
    for image_path, label in data:
        image = cv2.imread(image_path, 0)  # Read the image as grayscale
        image = cv2.resize(image, (32, 32))  # Resize the image to a fixed size
        image = image.flatten()  # Flatten the image
        processed_data.append((image, label))
    return processed_data

train_data = preprocess_data(train_data)
val_data = preprocess_data(val_data)
```

```python
# Train the model
X_train, y_train = zip(*train_data)
X_val, y_val = zip(*val_data)

scaler = StandardScaler()
svm_model = make_pipeline(scaler, SVC(kernel='linear'))
svm_model.fit(X_train, y_train)
```
